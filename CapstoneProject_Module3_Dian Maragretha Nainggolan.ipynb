### Metric Evaluation

Untuk mengevaluasi performa model klasifikasi, kita menggunakan **confusion matrix** yang terdiri dari empat komponen utama:

|                         | Prediksi: Tidak Churn (0) | Prediksi: Churn (1) |
|-------------------------|---------------------------|----------------------|
| **Aktual: Tidak Churn (0)** | TN (True Negative)          | FP (False Positive)  |
| **Aktual: Churn (1)**       | FN (False Negative)         | TP (True Positive)   |

**Penjelasan:**
- **TP (True Positive):** Pelanggan yang churn dan diprediksi churn.
- **TN (True Negative):** Pelanggan yang tidak churn dan diprediksi tidak churn.
- **FP (False Positive):** Pelanggan yang tidak churn tetapi diprediksi churn.
- **FN (False Negative):** Pelanggan yang churn tetapi diprediksi tidak churn.

Jenis Error & Konsekuensi Bisnis

| Jenis Error        | Definisi                                        | Dampak Bisnis                                                                 |
|--------------------|-------------------------------------------------|--------------------------------------------------------------------------------|
| **False Positive** | Pelanggan tidak churn, tapi diprediksi churn    | Biaya intervensi/retensi tidak perlu, margin berkurang                        |
| **False Negative** | Pelanggan churn, tapi diprediksi tidak churn    | **Pelanggan hilang tanpa dicegah, kehilangan pendapatan, biaya akuisisi baru** |


Evaluasi performa model dilakukan menggunakan lima metrik utama berikut:

1. **Accuracy**  
   Mengukur proporsi prediksi yang benar terhadap seluruh data.  
   Rumus:  
   $$
   \text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN}
   $$
   Metrik ini menggambarkan performa umum model, namun kurang representatif jika data tidak seimbang (misalnya, kasus churn dengan kelas positif yang sedikit).

2. **Precision**  
   Menunjukkan seberapa tepat prediksi churn yang dilakukan model, yaitu dari seluruh prediksi churn, berapa yang benar-benar churn.  
   Rumus:  
   $$
   \text{Precision} = \frac{TP}{TP + FP}
   $$
   Metrik ini penting untuk mengukur efisiensi penargetan agar tidak terlalu banyak pelanggan non-churn yang salah sasaran.

3. **Recall**  
   Mengukur seberapa banyak pelanggan yang benar-benar churn berhasil ditangkap oleh model.  
   Rumus:  
   $$
   \text{Recall} = \frac{TP}{TP + FN}
   $$

   > Dalam konteks churn prediction, **recall menjadi metrik utama**, karena:
   > - False Negative (gagal mendeteksi pelanggan yang akan churn) menyebabkan **kehilangan pelanggan**.
   > - Kehilangan pelanggan umumnya **lebih mahal** daripada mengakuisisi pelanggan baru, karena:
   >   - Kehilangan pendapatan berulang (lifetime value),
   >   - Biaya iklan dan promosi untuk menarik pelanggan baru,
   >   - Hilangnya kepercayaan terhadap brand.

4. **F1 Score**  
   Merupakan rata-rata harmonik dari precision dan recall.  
   Rumus:  
   $$
   \text{F1} = 2 \cdot \frac{\text{Precision} \cdot \text{Recall}}{\text{Precision} + \text{Recall}}
   $$
   Digunakan untuk menyeimbangkan antara kemampuan mendeteksi churn dan ketepatan prediksi.

5. **ROC AUC (Area Under Curve)**  
   Mengukur kemampuan model untuk membedakan antara pelanggan yang churn dan tidak churn pada berbagai ambang batas probabilitas.  
   Nilai AUC berkisar antara 0.5 (setara tebak acak) hingga 1.0 (model sempurna).

```python
for col in df.select_dtypes(include='object').columns:
    if col == 'Churn': continue
    vc = (df.groupby(col)['Churn'].value_counts(normalize=True)
          .rename('pct').reset_index())
    if vc.empty: continue
    churn_pct = (vc.pivot(index=col, columns='Churn', values='pct')
                 .fillna(0).mul(100))
    churn_pct_formatted = churn_pct.round(2).astype(str) + ' %'
    sort_by = 'Yes' if 'Yes' in churn_pct.columns else churn_pct.columns[0]
    churn_pct_sorted = churn_pct.sort_values(by=sort_by, ascending=False)

    print(f"\n=== Persentase Churn berdasarkan {col} ===")
    display(churn_pct_formatted.loc[churn_pct_sorted.index])
    
    plt.figure(figsize=(12, 4))
    plt.suptitle(f"Distribusi {col}", fontsize=14, fontweight='bold')
    order = df[col].value_counts(dropna=False).index
    sns.countplot(data=df, x=col, hue='Churn', order=order, palette='Set2')
    plt.title(f"Distribusi Churn berdasarkan {col}")
    plt.xlabel(col); plt.ylabel("Jumlah (Count)")
    plt.xticks(rotation=30, ha='right')
    plt.tight_layout()
    plt.show()
```

```python
rows = []
for col in df.columns:
    if col == 'Churn': 
        continue
    if df[col].dtype in ['int64', 'float64']:
        churned = df[df['Churn'] == 'Yes'][col].dropna()
        stayed = df[df['Churn'] == 'No'][col].dropna()
        if len(churned) > 5 and len(stayed) > 5:
            t, p = ttest_ind(churned, stayed, equal_var=False)
        else:
            p = np.nan
        rows.append([col, 'numeric', p])
    else:
        try:
            ct = pd.crosstab(df[col], df['Churn'])
            if ct.shape[0] >= 2 and ct.shape[1] >= 2:
                _, p, _, _ = chi2_contingency(ct)
            else:
                p = np.nan
        except:
            p = np.nan
        rows.append([col, 'categorical', p])

res = pd.DataFrame(rows, columns=['feature', 'type', 'p_value'])
res['selected'] = np.where(res['p_value'] < 0.05, 'Dipakai', 'Tidak Dipakai')
res = res.sort_values('p_value')
print(res)
```

**Standard Scaler**

Scaler digunakan untuk menstandarkan fitur numerik agar berada pada skala yang sama. Hal ini penting karena beberapa algoritma machine learning sensitif terhadap perbedaan rentang nilai antar fitur.  

**Standard Scaler** bekerja dengan mengubah setiap nilai fitur menggunakan rumus:

$$
z = \frac{x - \mu}{\sigma}
$$

di mana:  
- \(x\) adalah nilai asli  
- \(mu\) adalah rata-rata (mean) fitur  
- \(sigma\) adalah standar deviasi fitur  


```python
num_cols = X_train.select_dtypes(include=[np.number]).columns.tolist()
scaler = StandardScaler()
X_train_num_scaled = scaler.fit_transform(X_train[num_cols])
X_val_num_scaled   = scaler.transform(X_val[num_cols])
X_test_num_scaled  = scaler.transform(X_test[num_cols])
```

**Encoder dan One-Hot Encoder**

Encoder digunakan untuk mengubah data kategorikal menjadi bentuk numerik agar dapat diproses oleh model machine learning. Salah satu metode yang umum digunakan adalah **One-Hot Encoder**.  

**One-Hot Encoder** bekerja dengan membuat kolom baru untuk setiap kategori unik dan memberi nilai **1** pada kategori yang sesuai, serta **0** pada lainnya.  
Secara sederhana, jika sebuah fitur memiliki tiga kategori, misalnya:  
`["A", "B", "C"]`,  
maka representasi *One-Hot* akan menjadi:

| Kategori | A | B | C |
|-----------|---|---|---|
| A         | 1 | 0 | 0 |
| B         | 0 | 1 | 0 |
| C         | 0 | 0 | 1 |

Rumus umum transformasi One-Hot dapat dituliskan sebagai:

$$
x_{ij} = 
\begin{cases}
1, & \text{jika data ke-}i \text{ memiliki kategori } j \\
0, & \text{lainnya}
\end{cases}
$$


```python
cat_cols = X_train.select_dtypes(exclude=[np.number]).columns.tolist()
if int(sklearn.__version__.split(".")[1]) >= 4:  # scikit-learn 1.4+
    ohe = OneHotEncoder(sparse_output=False, handle_unknown="ignore")
else:
    ohe = OneHotEncoder(sparse=False, handle_unknown="ignore")


X_train_cat_enc = ohe.fit_transform(X_train[cat_cols])
X_val_cat_enc   = ohe.transform(X_val[cat_cols])
X_test_cat_enc  = ohe.transform(X_test[cat_cols])
# Gabungkan
X_train_enc = np.hstack([X_train_num_scaled, X_train_cat_enc])
X_val_enc   = np.hstack([X_val_num_scaled,   X_val_cat_enc])
X_test_enc  = np.hstack([X_test_num_scaled,  X_test_cat_enc])

print(f"Shape after encode: train={X_train_enc.shape}, val={X_val_enc.shape}, test={X_test_enc.shape}")

```

## Modeling & Evaluation

Tahap modeling merupakan proses membangun model machine learning menggunakan data yang telah melalui tahap preprocessing. Pada tahap ini, algoritma dilatih untuk mengenali pola antara fitur dan target agar dapat melakukan prediksi pada data baru.  

Setelah model terbentuk, dilakukan tahap **evaluation** untuk menilai seberapa baik model dalam memprediksi. Evaluasi biasanya menggunakan data validasi atau data test dengan berbagai metrik seperti akurasi, presisi, recall, atau f1-score.  


```python
def eval_all_metrics(y_true, y_pred, y_proba=None):
    """Kembalikan dict metrik lengkap."""
    out = {
        "accuracy": accuracy_score(y_true, y_pred),
        "precision": precision_score(y_true, y_pred, zero_division=0),
        "recall": recall_score(y_true, y_pred, zero_division=0),
        "f1": f1_score(y_true, y_pred, zero_division=0)
    }
    if y_proba is not None:
        try:
            out["roc_auc"] = roc_auc_score(y_true, y_proba)
        except Exception:
            out["roc_auc"] = np.nan
    else:
        out["roc_auc"] = np.nan
    return out

def get_proba(model, X):
    """Ambil probabilitas kelas positif bila ada; jika tidak, pakai decision_function dan skala ke [0,1]."""
    if hasattr(model, "predict_proba"):
        return model.predict_proba(X)[:, 1]
    elif hasattr(model, "decision_function"):
        s = model.decision_function(X)
        s = (s - s.min()) / (s.max() - s.min() + 1e-9)
        return s
    else:
        return None
```

Fungsi `eval_all_metrics()` digunakan untuk menghitung metrik utama seperti accuracy, precision, recall, f1-score, dan roc_auc jika tersedia probabilitas prediksi.  
Sementara `get_proba()` berfungsi untuk mengambil probabilitas kelas positif dari model, atau menyesuaikan skornya ke rentang [0,1] bila hanya tersedia `decision_function`.  
Kedua fungsi ini membantu menilai performa model secara konsisten.

### Penanganan Imbalance

Tahap ini dilakukan untuk mengatasi ketidakseimbangan kelas antara pelanggan yang churn dan tidak churn.  
Beberapa metode *resampling* digunakan hanya pada data **train**, kemudian semua model diuji ulang menggunakan data **validation**.  
Metode yang digunakan meliputi:

| Metode | Penjelasan Singkat |
|:--------|:------------------|
| **None** | Tanpa penyeimbangan data (baseline). |
| **RandomOverSampler** | Menambah data kelas minoritas dengan duplikasi acak. |
| **SMOTE** | Membuat sampel sintetis untuk kelas minoritas berdasarkan tetangga terdekat. |
| **SMOTEENN** | Kombinasi SMOTE dan Edited Nearest Neighbors (menambah minoritas dan membersihkan mayoritas). |
| **SMOTETomek** | Kombinasi SMOTE dan Tomek Links (menyeimbangkan sekaligus menghapus data tumpang tindih). |


```python
resamplers = {
    "None": None,
    "RandomOver": RandomOverSampler(random_state=seed),
    "SMOTE": SMOTE(random_state=seed),
    "SMOTEENN": SMOTEENN(random_state=seed),
    "SMOTETomek": SMOTETomek(random_state=seed),
}

imbalance_results = [] 
print("\n=== Benchmarking dengan Penanganan Imbalance (train-resampled) ===")
for rname, res in resamplers.items():
    if rname == "None":
        Xtr, ytr = X_train_enc, y_train
    else:
        Xtr, ytr = res.fit_resample(X_train_enc, y_train)
    for name, clf in models.items():
        try:
            clf.fit(Xtr, ytr)
            y_val_pred  = clf.predict(X_val_enc)
            y_val_proba = get_proba(clf, X_val_enc)
            mets = eval_all_metrics(y_val, y_val_pred, y_val_proba)
            imbalance_results.append((rname, name, mets))
            print(f"{rname:<11} | {name:<12} | recall={mets['recall']:.4f} | f1={mets['f1']:.4f}")
        except Exception as e:
            print(f"{rname:<11} | {name:<12} | ERROR: {e}")

imbalance_sorted = sorted(imbalance_results, key=lambda x: x[2]["recall"], reverse=True)
best_resampler_name, best_model_name, best_val_metrics = imbalance_sorted[0]
print(f"\nBest after Imbalance: {best_model_name} + {best_resampler_name} | Recall Val={best_val_metrics['recall']:.4f}")
```

Setelah dilakukan penyeimbangan, nilai recall meningkat signifikan di hampir semua model. Metode SMOTEENN memberikan hasil terbaik dengan model QDA, menghasilkan recall tertinggi sebesar 0.8682. Hal ini menunjukkan bahwa kombinasi tersebut paling efektif dalam mengenali pelanggan churn. Secara keseluruhan, penanganan imbalance membuat model lebih sensitif terhadap kelas churn, sehingga dapat membantu perusahaan lebih proaktif dalam mendeteksi pelanggan yang berpotensi berhenti berlangganan.

```python
def build_local_grid(model_name, best_params, pct_span=0.35, num_points=7):

    def _is_bool(x):
        import numpy as _np
        return isinstance(x, (bool, _np.bool_))

    # nama parameter yang harus berada di [0,1]
    prob_like = {
        "reg_param", "subsample", "colsample_bytree", "colsample_bylevel",
        "colsample_bynode", "max_samples", "max_features",
        "learning_rate", "l1_ratio", "bagging_temperature", "random_strength",
        "subsample_freq", "subsample_for_bin"  # note: freq bukan prop, tapi biarkan tetap aman
    }

    local = {}
    for k, v in best_params.items():
        # 1) Boolean harus duluan (bool is subclass of int)
        if _is_bool(v):
            local[k] = [bool(v)]
            continue

        # 2) String/categorical: pertahankan
        if isinstance(v, str):
            local[k] = [v]
            continue

        # 3) QDA.reg_param: clamp [0,1] dan perluas kecil
        if model_name == "QDA" and k == "reg_param":
            base = float(v)
            arr = np.linspace(max(0.0, base * (1 - pct_span)),
                              min(1.0, base * (1 + pct_span)),
                              num_points)
            # tambahkan tetangga +-0.05 untuk jaga-jaga
            extras = [max(0.0, base - 0.05), min(1.0, base + 0.05)]
            vals = np.clip(np.concatenate([arr, extras, [base]]), 0.0, 1.0)
            local[k] = sorted(set(float(x) for x in vals))
            continue

        # 4) Integer: tetangga sekitar
        if isinstance(v, (int, np.integer)):
            neigh = sorted({max(1, int(v) + d) for d in [-3, -2, -1, 0, 1, 2, 3]})
            local[k] = neigh
            continue

        # 5) Float umum: linspace sekitar nilai, clamp bila perlu
        if isinstance(v, (float, np.floating)):
            base = float(v)
            lo = base * (1 - pct_span)
            hi = base * (1 + pct_span)
            if k in prob_like:
                lo, hi = 0.0, 1.0
                # jika base dekat 0 atau 1, geser sedikit supaya ada variasi
                if base <= 0.0:
                    base = 0.05
                if base >= 1.0:
                    base = 0.95
                arr = np.linspace(max(0.0, base - pct_span), min(1.0, base + pct_span), num_points)
            else:
                arr = np.linspace(max(1e-12, lo), max(1e-12, hi), num_points)
            vals = np.array(arr, dtype=float)
            local[k] = sorted(set(float(x) for x in vals if np.isfinite(x)))
            continue

        # 6) fallback: pertahankan nilai
        local[k] = [v]

    return local
```

## Quadratic Discriminant Analysis (QDA)

**QDA** adalah algoritma klasifikasi berbasis **teorema Bayes** yang mengasumsikan setiap kelas memiliki distribusi **Gaussian** dengan kovarians yang berbeda.  
Model ini menentukan kelas dengan probabilitas terbesar bahwa suatu data termasuk ke dalam kelas tersebut, menggunakan rumus:  

$$
\hat{y} = \underset{k}{\arg\max} \; P(Y = k | X = x)
$$

Fungsi diskriminan QDA dituliskan sebagai:  

$$
\delta_k(x) = -\frac{1}{2} \ln |\Sigma_k| - \frac{1}{2}(x - \mu_k)^T \Sigma_k^{-1}(x - \mu_k) + \ln \pi_k
$$

**Keterangan:**  
- $\mu_k$ : rata-rata fitur pada kelas ke-$k$  
- $\Sigma_k$ : matriks kovarians pada kelas ke-$k$  
- $\pi_k$ : probabilitas prior untuk kelas ke-$k$  

Untuk mencegah overfitting, QDA menggunakan regularisasi pada kovarians:  

$$
\Sigma_k' = (1 - \lambda)\Sigma_k + \lambda I
$$

dengan $\lambda = 0.8$, yang membantu menjaga kestabilan model dan mengurangi noise pada data hasil resampling.  

Pada kasus **churn**, QDA digunakan untuk memprediksi apakah pelanggan akan **berhenti (churn)** atau **tetap berlangganan**, berdasarkan fitur seperti `tenure`, `MonthlyCharges`, dan `Contract`.  
Model ini efektif karena dapat menangkap perbedaan distribusi antar kelompok pelanggan — misalnya pelanggan dengan masa langganan pendek dan biaya tinggi cenderung lebih berisiko churn.  

Model dengan parameter terbaik (`tol=1e-6`, `reg_param=0.8`) menghasilkan **recall sebesar 0.92**, menunjukkan kemampuannya dalam mendeteksi pelanggan berisiko churn secara akurat.  
Dengan performa ini, perusahaan dapat melakukan **intervensi lebih dini** untuk mempertahankan pelanggan bernilai tinggi dan mengoptimalkan strategi retensi.

```python
num_names = list(num_cols)

def _safe_ohe_names(ohe, cat_cols):
    try:
        return list(ohe.get_feature_names_out(cat_cols))
    except Exception:
        try:
            return list(ohe.get_feature_names(cat_cols))
        except Exception:
            cat_names_fallback = []
            for c in cat_cols:
                levels = pd.Index(pd.concat([X_train[c], X_val[c], X_test[c]]).astype(str).unique())
                for lvl in levels[1:]:
                    cat_names_fallback.append(f"{c}_{lvl}")
            return cat_names_fallback

cat_names = _safe_ohe_names(ohe, cat_cols)
feat_names = num_names + cat_names

def _align_names(names, X_enc):
    if len(names) != X_enc.shape[1]:
        print(f"[Warning] Panjang feat_names ({len(names)}) != n_features ({X_enc.shape[1]}). Nama fitur akan di-autogen.")
        return [f"feat_{i}" for i in range(X_enc.shape[1])]
    return names

feat_names_test = _align_names(feat_names, X_test_enc)
feat_names_trv  = _align_names(feat_names, np.vstack([X_train_enc, X_val_enc]))

def aggregate_by_original_feature(importance_series: pd.Series, cat_cols, ohe, num_cols):
    mapping = {}
    try:
        ohe_names = _safe_ohe_names(ohe, cat_cols)
        for name in ohe_names:
            matched = None
            for c in cat_cols:
                if name.startswith(c + "_") or name.startswith(c + "__"):
                    matched = c
                    break
            mapping[name] = matched if matched is not None else name
    except Exception:
        for name in importance_series.index:
            base = name.split("_")[0]
            mapping[name] = base

    agg = {}
    for n in num_cols:
        if n in importance_series.index:
            agg[n] = importance_series[n]
    for oname, orig in mapping.items():
        if oname in importance_series.index:
            agg.setdefault(orig, 0.0)
            agg[orig] += float(importance_series[oname])

    return pd.Series(agg).sort_values(ascending=False)

perm_test = permutation_importance(
    final_model,
    X_test_enc, y_test,
    scoring="roc_auc",
    n_repeats=30,
    random_state=seed,
    n_jobs=-1
)
pi_test = pd.Series(perm_test.importances_mean, index=feat_names_test).sort_values(ascending=False)
pi_test_std = pd.Series(perm_test.importances_std, index=feat_names_test).loc[pi_test.index]

X_trv = np.vstack([X_train_enc, X_val_enc])
y_trv = pd.concat([pd.Series(y_train), pd.Series(y_val)], ignore_index=True).to_numpy()

perm_trv = permutation_importance(
    final_model,
    X_trv, y_trv,
    scoring="roc_auc",
    n_repeats=30,
    random_state=seed,
    n_jobs=-1
)
pi_trv = pd.Series(perm_trv.importances_mean, index=feat_names_trv).sort_values(ascending=False)
pi_trv_std = pd.Series(perm_trv.importances_std, index=feat_names_trv).loc[pi_trv.index]

pi_test_agg = aggregate_by_original_feature(pi_test, cat_cols, ohe, num_cols)
pi_trv_agg  = aggregate_by_original_feature(pi_trv,  cat_cols, ohe, num_cols)

TOPN = 30

plt.figure(figsize=(16, 7))
pi_test.head(TOPN).plot(kind="bar")
plt.title(f"Permutation Importance (Transformed) — TEST (Top {TOPN}) [QDA]")
plt.ylabel("Mean decrease in ROC AUC")
plt.tight_layout()
plt.show()

plt.figure(figsize=(16, 7))
pi_trv.head(TOPN).plot(kind="bar")
plt.title(f"Permutation Importance (Transformed) — TRAIN+VAL Original (Top {TOPN}) [QDA]")
plt.ylabel("Mean decrease in ROC AUC")
plt.tight_layout()
plt.show()

plt.figure(figsize=(16, 7))
pi_test_agg.head(TOPN).plot(kind="bar")
plt.title(f"Permutation Importance (Aggregated by Original Feature) — TEST (Top {TOPN}) [QDA]")
plt.ylabel("Mean decrease in ROC AUC")
plt.tight_layout()
plt.show()

plt.figure(figsize=(16, 7))
pi_trv_agg.head(TOPN).plot(kind="bar")
plt.title(f"Permutation Importance (Aggregated by Original Feature) — TRAIN+VAL Original (Top {TOPN}) [QDA]")
plt.ylabel("Mean decrease in ROC AUC")
plt.tight_layout()
plt.show()

pi_test.to_csv("perm_importance_test_transformed.csv")
pi_trv.to_csv("perm_importance_trainval_transformed.csv")
pi_test_agg.to_csv("perm_importance_test_aggregated.csv")
pi_trv_agg.to_csv("perm_importance_trainval_aggregated.csv")
print("Saved permutation importance CSVs.")
```

Berdasarkan hasil **Permutation Importance** dari model **QDA**, terlihat bahwa fitur yang paling berpengaruh dalam memprediksi churn adalah **tenure** dan **MonthlyCharges**. Nilai *mean decrease in ROC AUC* pada kedua fitur ini jauh lebih tinggi dibandingkan fitur lainnya, menunjukkan bahwa durasi pelanggan berlangganan dan besarnya tagihan bulanan menjadi faktor utama yang memengaruhi kemungkinan pelanggan berhenti.

Selain itu, fitur seperti **Contract**, **InternetService**, dan **Dependents** juga memberikan kontribusi penting, meski dengan pengaruh yang lebih kecil. ### **Implementasi Bisnis Model Prediksi Churn**

**Simulasi Kasus**

Perusahaan memiliki **10.000 pelanggan aktif**, dengan tingkat churn historis sebesar **25% (2.500 pelanggan)** per periode.
Biaya kampanye retensi (promosi, telemarketing, atau diskon) ke satu pelanggan adalah **$20**.

Model terbaik (**QDA + SMOTEENN**) memiliki performa:

* **Recall (Churn)** = 0.85
* **Precision (Churn)** = 0.49
* **Recall (Non-Churn)** ≈ 0.81

Artinya:

* Model mampu mengenali **85% pelanggan yang benar-benar akan churn**.
* Dari seluruh pelanggan yang diprediksi akan churn, **49% benar-benar churn**.
* Sekitar **81% pelanggan yang sebenarnya tidak churn** berhasil dihindari dari kampanye yang tidak perlu.

---

**Tanpa Model (Kampanye ke Semua Pelanggan)**

| Komponen                     | Nilai                      |
| ---------------------------- | -------------------------- |
| Total pelanggan dihubungi    | 10.000                     |
| Biaya per pelanggan          | $20                        |
| **Total biaya kampanye**     | **$200.000**               |
| Pelanggan churn yang dicegah | 2.500                      |
| Biaya terbuang (tidak churn) | 7.500 × $20 = **$150.000** |
| **Total penghematan**        | $0                         |

---

**Dengan Model (Kampanye Berdasarkan Prediksi)**

| Komponen                                           | Perhitungan           | Nilai                 |
| -------------------------------------------------- | --------------------- | --------------------- |
| Pelanggan diprediksi churn                         | (2.500 / 0.49) × 0.85 | ≈ **4.300 pelanggan** |
| True Positive (benar-benar churn)                  | 0.85 × 2.500          | **2.125 pelanggan**   |
| False Positive (tidak churn tapi diprediksi churn) | 4.300 – 2.125         | **2.175 pelanggan**   |
| False Negative (churn tapi tidak terdeteksi)       | 2.500 – 2.125         | **375 pelanggan**     |
| **Total biaya kampanye**                           | 4.300 × $20           | **$86.000**           |
| **Biaya terbuang**                                 | 2.175 × $20           | **$43.500**           |

---

**Perbandingan Efisiensi**

| Komponen                     | Tanpa Model | Dengan Model | Penghematan  |
| ---------------------------- | ----------- | ------------ | ------------ |
| Total pelanggan dihubungi    | 10.000      | 4.300        | **5.700**    |
| Biaya total kampanye         | $200.000    | $86.000      | **$114.000** |
| Pelanggan churn yang dicegah | 2.500       | 2.125        | -375         |
| Biaya terbuang               | $150.000    | $43.500      | **$106.500** |

---

**Kesimpulan Bisnis**

Dengan menerapkan model prediksi churn:

* Perusahaan dapat **menghemat biaya kampanye hingga $114.000 per periode**.
* Tetap mampu **mendeteksi 85% pelanggan yang berpotensi churn**, menjaga efektivitas intervensi.
* Biaya kampanye menjadi **lebih efisien**, karena hanya pelanggan dengan risiko churn tinggi yang dihubungi.
* Strategi ini memungkinkan perusahaan untuk **memusatkan sumber daya pada pelanggan bernilai tinggi** tanpa mengorbankan terlalu banyak peluang retensi.
